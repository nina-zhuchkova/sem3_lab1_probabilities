```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
matplotlib.rcParams.update({'font.size': 13})
matplotlib.rcParams['lines.linewidth'] = 2

light_green = (0.4, 0.8, 0.2)
dark_purple = (0.4, 0.3, 0.7)
raspberry_pink = (1, 0, 0.5)
lignt_blue = (0.7, 0.7, 0.9)
```

```python
bonus = pd.read_csv("../files/BonusDice_probability.csv")
penalty = pd.read_csv("../files/PenaltyDice_probability.csv")
ordinary = pd.read_csv("../files/Dice_probability.csv")
three = pd.read_csv("../files/ThreeDicePool_probability.csv")
double_dice = pd.read_csv("../files/DoubleDice_probability.csv")
```

```python
plt.title("100-sided dice probability distribution")
plt.grid(c = (0.85, 0.85, 0.85), ls = '--', lw = 1, zorder = 0)
plt.bar(bonus["Value"], 100*bonus[" Probability"], 
        color = light_green, 
        width = 1, 
        edgecolor = dark_purple,
        linewidth = 1,
        label = "Bonus Dice",
        alpha = 0.5,
        zorder = 5)
plt.bar(penalty["Value"], 100*penalty[" Probability"], 
        color = raspberry_pink, 
        width = 1, 
        edgecolor = dark_purple,
        linewidth = 1,
        label = "Penalty Dice",
        alpha = 0.5,
        zorder = 5)
plt.bar(ordinary["Value"], 100*ordinary[" Probability"], 
        color = lignt_blue, 
        width = 1, 
        edgecolor = dark_purple,
        linewidth = 1,
        label = "Ordinary Dice",
        alpha = 0.5,
        zorder = 5)
plt.xlabel("Value")
plt.ylabel("Probability, %")
plt.legend()
plt.savefig("Ordinary_Bonus_Penalty_probability.png", dpi = 1000)
```

```python
plt.title("Three 6-sided dices sum probability distribution")
plt.grid(c = (0.85, 0.85, 0.85), ls = '--', lw = 1, zorder = 0)
plt.bar(three["Value"], 100*three[" Probability"], 
        color = light_green, 
        width = 1, 
        edgecolor = dark_purple,
        linewidth = 1,
        alpha = 0.5,
        zorder = 5)
plt.xlabel("Value")
plt.ylabel("Probability, %")
plt.xticks(np.linspace(0, 18, 10))
plt.savefig("Three_Dice_probability.png", dpi = 1000)
```

```python
plt.title("Bonus and Penalty dices sum probability distribution")
plt.grid(c = (0.85, 0.85, 0.85), ls = '--', lw = 1, zorder = 0)
plt.bar(double_dice["Value"], 100*double_dice[" Probability"], 
        color = light_green, 
        width = 1, 
        edgecolor = dark_purple,
        linewidth = 0.5,
        alpha = 0.5,
        zorder = 5)
plt.xlabel("Value")
plt.ylabel("Probability, %")
plt.savefig("Double_Dice_probability.png", dpi = 1000)
```
